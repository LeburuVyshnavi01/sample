```python
def generate_invoice(customer, items):
    print(f"Invoice for {customer}")
    total = 0
    for item in items:
        name, price = item
        print(f"{name}: ₹{price}")
        total += price
    return f"Total Amount: ₹{total}"

items = [("Book", 250), ("Pen", 50)]
print(generate_invoice("Ravi", items))
```

```python
# task 1: Bank Account System
# Objective:
# Create a class BankAccount with the following features:
# Attributes:
# account_holder (str)
# balance (float)
# account_number (int)
# Methods:
# deposit(amount) → Adds amount to balance
# withdraw(amount) → Subtracts amount if sufficient balance
# get_balance() → Returns current balance
# _str_() → Nicely formatted account summary

class BankAccount:
    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Holder: {self.account_holder}\nAccount Number: {self.account_number}\nBalance: ₹{self.balance}"

# Example usage
a = BankAccount("Ravi", 123, 1000)
a.deposit(200)
a.withdraw(100)
print(a.get_balance())
print(a)
```

```python
# task 2: Library Book Management System
# Objective:
# Manage individual books in a library with their availability status.
# Class: Book
# Attributes:
# title (str)
# author (str)
# isbn (str)
# available (bool) — default to True
# Methods:
# checkout() → Set available = False if book is available
# return_book() → Set available = True
# _str_() → Return a string like: "Title by Author (ISBN) - Available"

class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = True  # default value

    def checkout(self):
        if self.available:
            self.available = False

    def return_book(self):
        self.available = True

    def __str__(self):
        if self.available:
            return f"{self.title} by {self.author} ({self.isbn}) - Available"
        else:
            return f"{self.title} by {self.author} ({self.isbn}) - Not Available"


# Example usage
b = Book("Python Basics", "John Smith", "1234567890")

print(b)           # Book is available
b.checkout()       # Book is checked out
print(b)           # Book is not available
b.return_book()    # Book returned
print(b)           # Book is available again
```

```python
# Task : create a class ATM that:
# Keeps the balance private.
# Uses getter to check the balance.
# Uses setter (method) to deposit and withdraw money with validations

class ATM:
    def __init__(self):
        self.__balance = 0  # Private balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance = self.__balance + amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance = self.__balance - amount

# Example usage
atm = ATM()

atm.deposit(1000)
atm.withdraw(300)
print("Balance:", atm.get_balance())  # Output: 700
```

```python
# Inheritance 
# Task:
# Create a base class Employee with attributes name and salary.
# Then create two child classes:
# Developer → has an extra method code()
# Manager → has an extra method manage()
# Demonstrate that both can access name and salary from the parent.


class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Developer(Employee):
    def code(self):
        return f"{self.name} is coding."

class Manager(Employee):
    def manage(self):
        return f"{self.name} is managing the team."

# Usage demonstration
developer = Developer("Alice", 80000)
manager = Manager("Bob", 90000)

print(developer.name, developer.salary)  
print(developer.code())                  

print(manager.name, manager.salary)       
print(manager.manage())                
```

```python
# # Abstraction 
# Task:
# Create an abstract class Payment with an abstract method pay(amount).
# Then create two classes:
# CreditCardPayment → implements payment by card
# UPIPayment → implements payment by UPI
# Both must override the pay() method.


from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class UPIPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using UPI")

# Usage
card = CreditCardPayment()
upi = UPIPayment()

card.pay(100)  # Outputs: Paid 100 using Credit Card
upi.pay(200)   # Outputs: Paid 200 using UPI
```

```python
# Polymorphism 
# Task:
# Create classes Instagram, Facebook, and Twitter each with a method share_post().
# Write a function that accepts any social media object and calls share_post() without knowing which platform it is.

class Instagram:
    def share_post(self):
        print("Sharing on Instagram")

class Facebook:
    def share_post(self):
        print("Sharing on Facebook")

class Twitter:
    def share_post(self):
        print("Sharing on Twitter")

def share_any(platform):
    platform.share_post()

# Demonstration
for platform in (Instagram(), Facebook(), Twitter()):
    share_any(platform)
```

```python
# write a code output like this 5,4,3,2,1
class countdown:
    def __init__(self, start):
        self.current = start
    def __iter__(self):
        return self
    def __next__(self):
       if self.current <= 0:
          raise StopIteration
       num = self.current
       self.current -= 1
       return num

for num in countdown(5):
    print(num)
```

```python
# write a code output like this 1,2,3,4,5
class countup:
    def __init__(self, end):
        self.current = 1
        self.end = end
    def __iter__(self):
        return self
    def __next__(self):
       if self.current > self.end:
          raise StopIteration
       num = self.current
       self.current += 1
       return num

for num in countup(5):
    print(num)
```

```python
n = 10
a , b = 0 ,1
print("Fibonacci series:")
for i in range(n):
    print(a,end=' ')
    a , b = b , a+b

# using recurstion fibonacci
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
    
for i in range(10):
    print(fibonacci(i),end=' ')
```

```python
n = 5
fact = 1
for i in range(1, n+1):
    fact = fact*i
print(fact)

# using recurstion factorial
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
print(fact(5))
```
